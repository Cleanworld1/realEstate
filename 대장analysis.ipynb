```python
import pandas as pd
import os

path_dir = 'C:/Users/skimf/Projects/realEstate/translated'

file_list = os.listdir(path_dir)

parameters_descriptions = []

for filename in file_list:
    df = pd.read_excel(path_dir + '/' + filename)
    
    list_a = df['출력명'].tolist()
    list_b = df['출력설명'].tolist()
    zipped_tuple = list(zip(list_a, list_b))

    for item in zipped_tuple:
        if len(parameters_descriptions)>0:
            x, y = map(list,zip(*parameters_descriptions))
        else:
            x = []
        if item[0] in x:
            continue
        else:
            parameters_descriptions.append(item)

```
